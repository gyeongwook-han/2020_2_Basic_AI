# 3. Analytic Geometry
## Manhattan Norm($l_1$ norm)
### R^2 평면에 원을 그릴 때, |x1|과 |x2|의 합이 n인 경우 
### $ ||x||_1 = \Sigma_{i=1}^{n} |x_i| $
## Euclidean Norm($l_2$ norm)
### R^2 평면에 원을 그릴 때, x1^2 과 x2^2의 합이 n^2인 경우
### $ ||x||_2 = \sqrt{ \Sigma_{i=1}^{n} {x_i}^2} $

## 실습에 필요한 Library Import

```python
import numpy as np
import matplotlib.pyplot as plt
# 그래프가 그려질 캔버스의 크기 정하기
plt.rcParams[ "figure.figsize" ] = (10, 10)
```

# 유클리디안 Norm, 원 그리기

```python
# 1. l_2 norm 그리기 = 유클리디안 Norm 그리기
xRight = np.linspace(0, 1, 50)
xLeft = np.linspace(-1, 0, 50)

xarr = [xRight, xLeft, xLeft, xRight]
xarr = np.array(xarr)
xarr = xarr.reshape(-1)

yarr = [np.sqrt(1-xRight**2), np.sqrt(1-xLeft**2), -np.sqrt(1-xLeft**2), -np.sqrt(1-xRight**2)]
yarr = np.array(yarr).reshape(-1)

plt.scatter(xarr, yarr, s = 5, color ='r')
plt.show()
```

# 맨하탄 Norm, 원 그리기

```python
# 2. l_1 norm 그리기 = 맨하탄 Norm 그리기
# x 범위 설정은 동일하다.
xarr = [xRight, xLeft,xLeft, xRight]
xarr = np.array(xarr)
xarr = xarr.reshape(-1)

yarr = [1-xRight, 1+xLeft, -(1+xLeft), -(1-xRight)]
yarr = np.array(yarr).reshape(-1)

plt.scatter(xarr, yarr,s=5, color ='b')
plt.show()
```

# 그래프에 범주와 제목을 함께 표시하기

```python
# 1. l_2 norm 그리기 = 유클리디안 Norm 그리기
xRight = np.linspace(0, 1, 50)
xLeft = np.linspace(-1, 0, 50)

xarr = [xRight, xLeft, xLeft, xRight]
xarr = np.array(xarr)
xarr = xarr.reshape(-1)

yarr = [np.sqrt(1-xRight**2), np.sqrt(1-xLeft**2), -np.sqrt(1-xLeft**2), -np.sqrt(1-xRight**2)]
yarr = np.array(yarr).reshape(-1)

# 유클리디안 Norm 그래프 점 찍기
plt.scatter(xarr, yarr, s = 15, color ='r')

# 2. l_1 norm 그리기 = 맨하탄 Norm 그리기
# x 범위 설정은 동일하다.
xarr = [xRight, xLeft,xLeft, xRight]
xarr = np.array(xarr)
xarr = xarr.reshape(-1)

yarr = [1-xRight, 1+xLeft, -(1+xLeft), -(1-xRight)]
yarr = np.array(yarr).reshape(-1)

# 맨하탄 Norm 그래프 점 찍기
plt.scatter(xarr, yarr,s=15, color ='b')

#제목
# ''  작은 따옴표 안에 있는 문자열이 제목이 된다.
plt.title('Manhattan Norm(L1, blue) , Euclidean Norm(L2, red)')
plt.legend(["l2", "l1"])

plt.show()
```
